```python
df.replace(r'^\s*$',np.nan, regex =True, inplace =True)
```

```python
df.columns = df.columns.str.lower()
```

```python
#filled null values with unknown, anonymous

df['region']=df['region'].fillna('unknown')
df['customer_name']= df['customer_name'].fillna('anonymous')
```

```python
df[['region','customer_name']].head(100)
```

```python
#dropped order_ids with no datas in it

df.dropna(subset=['order_id'], inplace=True)
```

```python
df.columns
```

```python
numeric_cols = ['price', 'discount(%)', 'quantity','rating', 'delivery_delay_days']
for col in numeric_cols:
    df[col]=pd.to_numeric(df[col], errors = 'coerce')
```

```python
df.describe()
```

```python
df.isnull().sum().sum()
```

```python
#changed dates formats and checked order date and delivery date 

df['order_date']=pd.to_datetime(df['order_date'], errors  = 'coerce')
df['delivery_date']=pd.to_datetime(df['delivery_date'], errors  = 'coerce')
mask  = df['order_date'] > df['delivery_date']
```

```python
df.loc[mask,'order_date'] = df['delivery_date']
```

```python
mask.sum()
```

```python
#checked whether delivery date > order date

mask  = df['order_date'] > df['delivery_date']
mask.sum()
```

```python
mask.sum()
```

```python
df.drop_duplicates(inplace = True)
```

```python
df.duplicated()
```

```python
df.columns
```

```python
# removed spaces and used title function

text_columns = ['region','category', 'payment_method','product_name']
for cols in text_columns:
    df[cols]= df[cols].str.strip().str.title()
```

```python
df['review_text'] =df['review_text'].str.strip()
df['customer_name'] =df['customer_name'].str.strip()
```

```python
df['customer_name']
```

```python
# removing str from price and converting it into float

df['price']=df['price'].astype(str).str.replace(r'[^\d\.]','',regex = True)
df['price']=pd.to_numeric(df['price'], errors = 'coerce')
```

```python
df
```

```python
# clean email format
df['customer_email']=df['customer_email'].apply(lambda x : x if isinstance(x,str) and '@' in x and '.' in x else np.nan)
```

```python
df.isnull().sum()
```

```python
df['product_name'].fillna('unnamed product', inplace= True)

```

```python
df['review_text'].fillna('No Review', inplace=True)
df['warehouse_id'].fillna('Unknown', inplace=True)
df['currency'].fillna('AED', inplace=True)
df['delivery_delay_days'].fillna(0, inplace=True)
df['customer_email'].fillna('unknown@email.com', inplace=True)
```

```python
df.fillna({'price': df['price'].median(),
           'quantity': df['quantity'].median(),
           'discount(%)': df['discount(%)'].median(),
          'rating' : df['rating'].median()},
          inplace=True)
```

```python
df['order_date'].fillna(pd.Timestamp('2023-01-01'), inplace=True)
df['delivery_date'].fillna(df['order_date'], inplace=True)
```

```python
df['return_requested'].fillna('no', inplace= True)
```

```python
numeric_cols = ['price', 'quantity', 'discount(%)', 'rating', 'delivery_delay_days']
for col in numeric_cols:
    cap = df[col].quantile(0.99)

    df[col] = np.where(df[col] > cap, cap, df[col])
```

```python
df['shipping_cost']=pd.to_numeric(df['shipping_cost'], errors = 'coerce')
```

```python
df['shipping_cost'].fillna(df['shipping_cost'].median(), inplace=True)
```

```python
df.describe()
```

```python
df.info()
```

```python
df['total_price'] = df['price'] * df['quantity']
```

```python
df['is_delayed'] = np.where(df['delivery_delay_days'] > 0, 'Yes', 'No')
```
